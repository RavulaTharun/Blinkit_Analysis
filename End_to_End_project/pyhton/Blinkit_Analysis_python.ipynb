# **BUSINESS REQUIREMENTS**

# **KPI's REQUIREMENTS**

```python
#Total Sales
total_sales = df['Sales'].sum()

#Average Sales
avg_sales = df['Sales'].mean()

#No of Items Sold.
no_of_items_sold = df['Sales'].count()

#Average Rating
avg_rating = df['Rating'].mean()

#Display

print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.1f}")
print(f"No of Items Sold: {no_of_items_sold:,.0f}")
print(f"Average Ratings: {avg_rating:,.1f}")
```

# **CHARTS REQUIREMENTS**

## **Total Sales by Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat,labels=sales_by_fat.index,autopct='%.0f%%',startangle=90)

plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show
```

## **Total Sales by Item Type**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
bars=plt.bar(sales_by_type.index,sales_by_type.values)


plt.title('Total Sales by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.xticks(rotation=-90)


for bar in bars :
    plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),
            f'{bar.get_height():,.0f}',ha='center',va='bottom',fontsize=8)

plt.tight_layout()
plt.show()
```

## **Fat Content by Outlet for Total Sales**

```python
sales_by_oulet_nd_item_fat= df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()

x=sales_by_oulet_nd_item_fat.plot(kind='bar',figsize=(8,5),title='Outlet Tier by Item Fat Content')    

plt.xlabel('Outlet Location Tier')
plt.legend(title='Item Fat Content')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

## **Total Sales by Outlet Establishment**

```python
sales_by_oulet_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(10,6))
x= plt.plot(sales_by_oulet_year.index,sales_by_oulet_year.values,marker='o',linestyle='-')

for x,y in zip(sales_by_oulet_year.index,sales_by_oulet_year.values):
    plt.text(x,y,f'{y:,.0f}', ha='center',va='bottom',fontsize=8)


plt.title('Tital Sales by Oulet Establishment Year')
plt.xlabel('Oulet year')
# plt.grid(True)
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

## **Percentage of Sales by Outlet Size**

```python
sales_by_oulet_size = df.groupby('Outlet Size')['Sales'].sum()


plt.figure(figsize=(4,4))
y=plt.pie(sales_by_oulet_size,labels=sales_by_oulet_size.index,autopct='%1.1f%%',startangle=90)


plt.title('Percentage of Sales by Outlet Size')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## **Sales by Outlet Location**

```python
sales_by_outlet_location = df.groupby('Outlet Location Type')['Sales'].sum()


d=sales_by_outlet_location.plot(kind='barh', figsize=(8,3), title='Total Sales by Outlet Location')


plt.ylabel('Outlet Location Type')
plt.xlabel('Total Sales')
plt.tight_layout()
plt.show()
```

## **All Metrics by Outlet Type**

```python
Total_sales = df.groupby('Outlet Location Type')['Sales'].sum().rename('Total Sales')
Avg = df.groupby('Outlet Location Type')['Sales'].mean().rename('Average Sales')
Avg_Rating = df.groupby('Outlet Location Type')['Rating'].mean().rename('Average Rating')
counts = df.groupby('Outlet Location Type')['Sales'].count().rename('Sales Count')

d = pd.concat([Total_sales, Avg, Avg_Rating, counts], axis=1)

# Normalize each column between 0 and 1
d_norm = (d - d.min()) / (d.max() - d.min())

# Plot heatmap with normalized data
sns.heatmap(d_norm, annot=d, fmt=".2f", cmap="YlGnBu")
plt.show()
```
